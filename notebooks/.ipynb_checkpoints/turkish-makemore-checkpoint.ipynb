```python
import csv

names = []
with open("female_name_tally", "r") as f:
    data = csv.reader(f)
    for row in data:
        names.append(row[0].split(" ")[0])
```

```python
names = [name.lower() for name in names]
```

```python
non_turkish_chars = {'-', '<', '>', '_', '?', '1', 'x', "'", '(', ')', '0', 'q', 'w', '`', '.'}
names = ['.' + name + '.' for name in names if not any(char in non_turkish_chars for char in name)]
```

```python
unique_counts = {}

for name in names:
    for char in name:
        unique_counts[char] = unique_counts.get(char, 0) + 1
```

```python
keys_list = list(unique_counts.keys())
itos = {}
stoi = {}

for i in range(len(keys_list)):
    char = keys_list[i]
    itos[i] = char     # Maps index -> character
    stoi[char] = i     # Maps character -> index
```

```python
len(keys_list)
```

```python
N = torch.zeros(24, 24) # number of unique characters in our dataset including our special . character

for name in names:
    for char1, char2 in zip(name, name[1:]):
        N[stoi[char1]][stoi[char2]] += 1
        
```

```python
row_sums = N.sum(1, keepdim=True)

P = (N + 1).float()
P /= P.sum(1, keepdim=True)
```

```python
import torch

g = torch.Generator().manual_seed(555)

for i in range(5):
    out = []
    ix = stoi['.']
    
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        if ix == stoi['.']:
            break
        out.append(itos[ix])
        
    print(''.join(out))
```

```python
log_likelihood = 0.0
n = 0 
for name in names:
    wrapped_name = list(name)
    
    for char1, char2 in zip(wrapped_name, wrapped_name[1:]):
        ix1 = stoi[char1]
        ix2 = stoi[char2]
        
        prob = P[ix1, ix2]
        
        log_prob = torch.log(prob)
        
        log_likelihood += log_prob
        n += 1


nll = -log_likelihood

average_nll = nll / n
print(f"Average NLL Loss: {average_nll.item():.4f}")
```
